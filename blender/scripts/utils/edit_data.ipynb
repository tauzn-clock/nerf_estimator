```python
import json
import os
```

```python
FILE_PATH = "../../data/jackal_irl_one_spin"
with open(os.path.join(FILE_PATH, "transforms.json"), "r") as f:
    data = json.load(f)
```

```python
new_frames = []
cnt = 0
default = data["frames"][0]

for filename in sorted(os.listdir(os.path.join(FILE_PATH, "images"))):
    file_path = os.path.join(os.path.join(FILE_PATH, "images"), filename)
    if os.path.isfile(file_path):
        #New Frame
        new_f = default.copy()
        new_f["file_path"] = "images/{0}.png".format(cnt)
        new_f["mask_path"] = "masks/{0}.png".format(cnt)
        print(new_f)
        new_frames.append(new_f)

        #Rename file
        os.rename(file_path, os.path.join(os.path.join(FILE_PATH, "images"), "{0}.png".format(cnt)))
            
        cnt += 1
```

```python
data["frames"] = new_frames

with open(os.path.join(FILE_PATH, "transforms.json"), "w") as f:
    json.dump(data, f, indent=4)
```
